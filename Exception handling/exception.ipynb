## EXCEPTION
- Exception are unexpected events or errors that occurs during the execution of program,disrupts the normal flow of program.
 print('start')
 print(10/0)        #raise ZeroDivisonError
 print('end')       #this line will never run

- ZeroDivivsonError is counted as exception n becoz of this next line will not be executed.
- we can handle the exception.

## EXCEPTION HANDLING
   keyword    ---->   Purpose
- try         ----> wrap the block of code that might couse error
- except      ----> handles exception if occurs
- else        ----> Run code only if no expcetion occurs.
- finally     ----> run code wheather there is an exception or not.
- raise       ----> Manually throw exception.


```python
a = int(input('enter the number'))
try:
    print (10/a)                     #at 0 it will create an exception/error(zeroDivisonError)
#try always run with except keyword
except Exception as err:             #err is variable
    print(f"sorry there is an error as {err}")
else:
    print('there is no exception')        #else will run if except don't run

finally:
    print('i will no matter what!!')     #it don't depend on except it will always run

print('ok i have done division')
```
