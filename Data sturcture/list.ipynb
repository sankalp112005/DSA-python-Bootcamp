# Lists

### lists are--
 - Mutable : Object value can be chnaged afte creation
 - Duplicates : list can have multiple same value of same type.
 - Ordered : list saves value in index or in order in which they are added.
 - Heterogenous : we can have multiple data type in list (i.e. int,string,bool,float,function).
 - List are created using square bracket[].


```python
fruits=["apple","banana","strwaberry",'mango']
nums=[1,5,6,7,34,69]
mix=[182,"sankalp","jain",8.55,11022005,True,print()] 
```

```python
#printing by index.
print(nums[4])    
print(mix[1])
```

list are indexed and sliced as that of string.
### SLICING of LIST

```python
print(nums[0:4])         #last number is not included
print(fruits[1:4])
print(mix[-2:-4:-1])
```

### LIST TRAVERSING

```python
lst = [11,23,56,78,90,76,45,23]
#1.using index    (most used way)
for i in range (len(lst)):
    print(lst[i])

#2.using direct   (when you concern only for vlaues not index)
for i in lst:
    print(i)
```

### METHODS IN LIST   --defined inside the class concept of oops

```python
numbers = [5, 2, 9, 1, 5, 6] # Initial list

numbers.append(18)                #Adds 10 to the end
numbers.insert(2, 15)             #Inserts 15 at index 2
numbers.extend([28, 25, 30])      #Adds multiple elements at the end
numbers.remove(5)                 #Removes the first occurrence of 5
popped_item= numbers.pop(3)       #Removes and stores the element at index 3
index = numbers.index(6)          #Finds the index of 6
count_5= numbers.count(5)         #Counts occurrences of 5
numbers.sort()                    #Sorts the list in ascending order
numbers.reverse()                 #Reverses the list order
new_numbers = numbers.copy()      #Creates a copy of the list
numbers.clear()                   #Removes all elements from the list
```

```python
#1.Print positive and negative lement of the list
list=[1,-4,76,-67,-66,34,10,-46]
print('positive elements are -')
for i in list:
    if i >= 0:
        print(i)
print('negative elements are- ')
for i in list:
    if i < 0:
        print(i)
```

```python
#2.MEAN OF LIST ELEMENT
list=[1,2,5,7,4,8,9,4]
sum=0
for i in list:           #mjhe value se matlb hai naki index se islie noy used index.
    sum+= i
mean = sum / len(list)
print(mean)
print(sum / len(list))
```

```python
#3.FIND GREATEST ELEMENT AND PRINT ITS INDEX
l=[123,45,333,789,987,345,236,754]
largest = l[0]
index = 0                 #manke chalo abhi pehla element hi sabse bada hai
for i in range (len(l)):
    if l[i]>largest:
        largest = l[i]     #update th etag of largest in each iteration
        index = i
print(f'your largest number is {largest} & its index is {index}')
```

```python
#4.FIND THE SECOND LARGEST NUMBER
l=[12,56,87,34,56,90]
largest = l[0]
sec_largest=l[0]
for i in l:
    if i>largest:
        sec_largest = largest
        largest = i
    elif i>sec_largest:
        sec_largest=i

print(largest,sec_largest)
```

```python
#5.CHECK IF LIST IS SORTED OR NOT
a=[12,13,14,15,16,17]
for i in range(len(a)-1):
    if a[i]<a[i+1]:
        continue
    else :
        print('your list is not sorted')
        break
else:
    print('your list is sorted')
```
