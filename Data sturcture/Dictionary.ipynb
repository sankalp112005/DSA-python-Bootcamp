# DICTIONARY
### DISCTIONARY IS-
- it has key and value pair.
- MUTABLE-can change,add ,remove key value pair
- DUPLICATE- keys must be UNIQUE,,but value can be DUPLICATE.
- ORDER - follow insertion order.
- HETEROGENOUS - can store different types of keys and values.
- d={1:"sankalp",2:182}        #1,2 are keys & sankalp,182 are value

```python
d = {1:'sankalp',4:182,6:'sankalp','hello':'hamilton'}
print(d)
print(d[6])           #key behave as index for dictionary
```

```python
#CRUD operation in dictionary
d = {1:100,2:200,3:300,4:400,5:500}
d.update({6:600})         #used to create/add new key nd value
d[7]=700
print(d)

d[3]=30           #key values is updated.
print(d)

del d[3]          #deleting
print (d)
```

### Travesing in dictionary

```python
d = {1:100,2:200,3:300,4:400,5:500}
for i in d.values():      #using method,gives values
    print(i)

for i in d:              
    print(i)             #gives keys
    print(d[i])          #gives value

```

#### DEEP COPY AND SHALLOW COPY

```python
d = {1:100,2:200,3:300,4:400}
#Deep Copy- make a copy in another dict and if we change in copy dict change is reflected in  main dict also.
d2=d    #d2 mai d copy krdi
d2[2]=6969

print(d)   #change in d2 reflected in d also

#Shallow Copy- makes another shallow copy and its chnage is upto that dict only.
#done with the help of copy() method
d3=d.copy()
d3[4]=2766       #change will be only in d3.
print(d3)      
```

##### question on dict

```python
#1.PYTHON SCRIPT TO MERGE TWO DICTIONARIES
d1={1:10,2:20,3:30}
d2={3:60,4:40,5:50}

for i in d2:
    d1[i]=d2[i]
print(d1)
```

```python
#2.SUM ALL THE VALUES IN DICTIONARY
d1={1:10,2:20,3:30}
sum=0
for i in d1.values():
    sum += i
print(sum)
```

```python
#3.COUNT FREQUENCY OF EACH ELEMENT in list
a= [1,2,3,1,4,2,1,3,4,1,2,4,5]
dict = {}                    #make dictioanry of frequencies
for i in a:
    if i in dict.keys():
        dict[i]+=1
    else:
        dict[i]=1              #new key create hogi 

print(dict)
```

```python
#4.COMBINE TWO DICT. BY ADDING VALUE OF COMMON KEYS
d1={1:10,2:20,3:30}
d2={3:60,4:40,5:50}
for i in d2:
    if i in d1.keys():
        d1[i] += d2[i]
    else:
        d1[i] = d2[i]
```
