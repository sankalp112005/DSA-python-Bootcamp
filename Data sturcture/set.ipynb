# SET
*set are-
- MUTABLE - change the value
- NO-DUPLICATE - every element must be unique in set.
- UNORDERED - cannot accessd by index value.
- SEMI - HETEROGENOUS -can store number,string,tuple in it.
- defined by curly bracket { }.
- VALUES IN SET ARE STORED IN FORM OF HASHING.
- Each value in set is hashed using the hash function in python (hash()).

```python
s = {1,4,5,8,3,9,10}
for i in s:           #it will give random sequence according to hash values
    print(i)
```

```python
#METHODS IN SET
S={1,2,3,4,5,6,7,8}

(S.add(9)        #adds an element to set
S.remove(6)      #remove 6 form set (raise dispute if not found)
S.discard(10)    #remove 10 no dispute if not found
S.pop()          #removes a random element
S.clear()        #removes all element
```

```python
#MORE METHODS
a={1,2,4,6,7}
b={7,8,9,2,4}

#1.UNION            #print all elemnt excluding the repaated ones
s=a.union(b)
s=a|b            #shortcut of union
print(s)

#2.INTERSECTION     #gives common partr in two set
s=a.intersection(b)
s=a&b
print(s)

#3.DIFFERENCE       #gives common in first set
s=a.difference(b)
k=b-a
print(s,k)

#4.SYMMETRIC DIFFERENCE    #common part hatao aur bache hue differenctial parts ko nikal lena 
s=a.symmetric_difference(b)
k=b^a
print(k)
```
