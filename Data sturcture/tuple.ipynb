# TUPLE 
- Tuple are--
- IMMUTABLE-values can't be changed
- DUPLICATE- can have duplicate value.
- ORDERED- can access via index value,
- HETEROGENOUS - can have differnt data types .
- written with parenthesis ().
- tuple are somwhat like strings.

```python
tup=(2,3,5,7,8,9,4,5)
tuple_2= (2,3,5,6,"sankalp",8.55,print())
print(tuple_2)
```

```python
#TRAVERSING IN TUPLE
for i in tup:      #accessing direct values not index
    print(i)

for i in range(len(tuple_2)):     #accessing values by the use of index
    print(tuple_2[i])
```

### METHODS IN TUPLE

```python
index=tup.index(8)   #find the index of passed value/index method
print(index)

count=tup.count(5)   #count the occurance of claue in tuple/count method
print(count)
```

```python
#TUPLE UNPACKING
a,b,c,d = (1,2,3,4)    #it will unpack the tuple and every variable will get one value
print(a,b,c,d)

a=(23)
print(type(a))         #it will unpack the single value into integer.

b=(11,) 
print(type(b))         #it state another value is also coming.
```
