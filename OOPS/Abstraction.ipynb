## ABSTRACTION
- Used in simplifying complex feature, focuses on essential feature and hide the unnecessary detail.
- used to define common interface for different subclasses.

#### ABSTRACT CLASSES AND METHODS
- ABSTRACT CLASSES are classes that contains one or more abstarct methods.

- a method that is defined but not implemented in the abstract class.subclasses must provide the implementation.

- in pyhton it is done by the library (abc module)
 
- make the rules that need to be followed in the class.

```python
from abc import ABC, abstractmethod

class abstract(ABC):
    @abstractmethod                 #decorator
    def perimeter (self):
        pass

    @abstractmethod
    def area(self):
        pass

class Square(abstract):             
    def __init__(self,side):
        self.side = side

    def perimeter(self):
        print("it is perimeter")

    def area(self):
        print("it is area")


class Circle(abstract):
    def __init__(self,radius):
        self.radius = radius

obj = Square(5)
obj2 = Circle(4)      #it will give error becoz abstract method not implemented
```

#### DUNDER METHODS
- Special method in pyhton sytart and end with double undrscore __ (__init__,__str__, __add__ etc)
 
- get automaticallly called when you do certain actions  on an objects

- they help in
* customize behaviour in class
* make your class ibject behave like built - in data types(like string ,lsits etc)

```python
class Animal:
    def __init__(self,name,age):
        self.name  =  name
        self.age = age

    def __str__(self):
        return f"hello how are you my name is {self.name}"

    def __add__(self,other):          #self is first thing other is second
        return f"your sum of ages are {self.age + other.age}"

obj = Animal('lion',21)
obj2 = Animal("chetah",20)

print(obj)               #we don't need to call function for print it prints the object
print(obj + obj2)
```
