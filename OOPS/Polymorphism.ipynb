## POLYMORPHISM
- as word mean 'many forms'./ same name having different forms.
- it allows the same interface or method name to behave differently depending on the object or context.
- python overwrite the function. means most recent(latest) will execute.
- it is achieved by 2 ways, compile time languages (c++,java) has 3 ways.
- python don't support METHOD OVERLOADING. 
* ways to achieve POLYMORPHISM -

##### 1.METHOD OVERRIDING
* This is where the child class overrides a method of the parent class. pyhton decide at runtimme which method to call, based on object type.

- in simple - if you have one parent class & one derived class, both have same name method ,and object is calling that method,,,then method will be called of derived class,and base class method will be OVERRIDE

```python
#METHOD OVERRIDING -

class Animal:
    def show(self):                  #'self' written becoz it is instance function
        print('hello im sankalp')

class Human(Animal):
    def show(self):
        print('how are u')

obj = Human()
obj.show()                     #it will show/call latest funtion

```

##### 2.DUCK TYPING
* Based on philosophy "if it walk like a duck and quack like a duck, it must be a duck."

* we have multiple class and multiple method with same name, but they are separately deifned by the class.

```python
class Bird:
    def fly(self):
        print('Bird is flying')
class Aeroplane:
    def fly(self):                   
        print('Aeroplane is flying')
#both same name method 'fly' will call in object separately.
obj = Bird()
obj2 = Aeroplane()

obj.fly()
obj2.fly()
```
