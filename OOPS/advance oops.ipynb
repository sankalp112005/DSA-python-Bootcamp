### DECORATOR

- decorator is just a FUNCTION that wraps and modifies the another FUNCTION without changing its actual code
- for crating decorator first we need to create decorator functions and then inside that we will create a wrapper.
- adds additional functionality before and after the original function execute

##### creating decorator

```python
def decorate(func):            #creating function of decorator
    def wrapper(a,b):             #wrapper function call
        #wrapper capture all the argument that was given in function
        print("the addition to your number are")
        func(a,b)                 #func is parameter, jis function ke uper decorate banya hai woh func mai argument jata h
        print("thank you!!")
    return wrapper             #wrpapper class will go to decorate 
                                
@decorate                       #any name we can give
def add(a,b):
    print(f"your total is {a+b}")

add(20,30)
```

#### Args and Kwargs
- special keyword in python used in function definition to accept a flexible number of arguments.

- we can use any name ahead of *,** not only args and kwargs 
- *args are used for multiple positional arguments
- **kwargs are used for multiple key word arguments 

- 'args becomes a tuple' and 'kwargs become a dictionary'

- they are used best when:
  - when you don't know how many inputs you'll get
  - help in building flexible function,APIs,decortors, and more 

```python
# *args
def addition(*args):         #args take the multiple argument in tuple format 
    sum = 0                  #used to capture multple argument when no. of argument is not ifx
    for i in args:
        sum += i 
    print(sum)


addition(12,13,14,15,45,7,8)    
```

```python
# **kwargs
def information(**kwargs):                 #used to capture multiple key value arguments
    print("your information is")
    for i in kwargs:
        print(f"{i} : {kwargs[i]}")

information(name = "sankalp",age = 21,collage = "lnct")
```

```python
# using both args & kwargs
def decorate(func):            #creating function of decorator
    def wrapper(*args,**kwargs):             #wrapper function call
        #wrapper capture all the argument that was given in function
        print("the addition to your number are")
        func(*args,**kwargs)                 #func is parameter, jis function ke uper decorate banya hai woh func mai argument jata h
        print("thank you!!")
    return wrapper             #wrpapper class will go to decorate 
                                
@decorate                       #any name we can give
def add(a,b,c):
    print(f"your total is {a+b}")

add(23,40,45)
```

### COMPREHENSION
- we can create list,dictionary in single line by using LOOPs and IF-ELSE statement. we don't need multiple line of code for this. 

###### make lsit for even in between 1 to 20

```python
l = []
for i in range(1,21):
    if i % 2 == 0:
        l.append(i)
print(l)

#doingin single line by COMPREHENSION
lst = [i for i in range (1,21) if i % 2!=0]
print(lst)

dic = {i : i**2 for i  in range (1,10)}      #doctipnary comprehension
print(dic)
```

### LAMBDA FUNCTION
- anonymous, inline function dedined using lambda keyword
- it is often used for short,simple fucntions that are used only once or temporarily.
- you can havee multiple argument but there will be only one expression

```python
addition = lambda a,b : a+b            #this is an object we need to print it
print(addition(2,5))

#even and odd
no = lambda a : "even" if a % 2 ==0 else "odd"   # we need to save lambda function in variable that act as object
print(no(5))
```

#### MAP
- Map is used for applying the function to multiple item.
- it take a List(or sequence)-
- applies the same funcntion on every item of list
- gives back a new list
- use map() when u want to trnasform every element in the list
 

```python
a = [1,2,3,4,5,6]
result = map(lambda x : x**2 , a)        #a gives the lsit to map
print(result)         #this will give a addreses
#we need to change the result into the list first
print(list(result))
```

#### FILTER
- it use to filter out the stuff.
- take a list
- checks each item using the function 
- keeps only the item that pass the test

```python
#normal method
def even(x):
    if x % 2 == 0:
        return True
    else:
        return False
a = [1,23,4,5,6,7,9,8,70,22]
result = filter(even,a)
print(list(result)) 

# lambda 
lst = [22,44,66,77,35,67,98]
result = filter(lambda x:True if x%2==0 else False , lst)
print(list(result))
```
