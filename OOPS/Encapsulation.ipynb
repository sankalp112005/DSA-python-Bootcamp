## ENCAPSULATION
- encapsulation means putting data (variables) and code(function) together in one place - inside the class.

- it also means hiding the internal details of things work,and show only what is needed. 

- it is usefull because -
1. it keeps data form being changed by mistake
2. it make code easy and clean
3. it gives control over what other can access or change. 

- encapsulation is done by access modifiers in python.

### ACCESS MODIFIERS

- it means how we give access of our attribute and methods to object or inherited classes. they are of three types.

1. PUBLIC ATTRIBUTE N METHOD
abhi tak jitne method and attrbite padhe sab public the..means inherited class and object can access them no matter

2. PROTECTED ATTRIBUTE N METHOD
- protected members are created using single undescore but it can still be accesd outside the class 

- python not use protected access like(c++ n java). but it uses the naming convemtion (gives idea to developer that this is protected method)

3. PRIVATE ATTRIBUTE N METHOD
- it cannot be accessed from outisde the class-only from inside the class where it is defined.
- we use twp underscore __ before the name to make it private 


```python
# PROTECTED ATTRIBUTE
class Factory:
    _a = 'pune'                   #_a is protected attribute
    
    def _show(self):              #_show is protected method
        print("i'm pune facotry")

class Bhopal(Factory):
    def show2(self):
        print(super()._a)

obj = Bhopal()
obj.show2()
```

```python
# PRIVATE ATTRIBUTE
class Factory:
    __a = 'pune'                   #__a is private koi bhi class or object inherit nhi kar sakta
    
    def show(self):              #_show is protected method
        print("i'm pune facotry")
obj = Factory()
obj.__a
print(obj.__a)
obj.__show()
```

```python
class Demo:
    def __init__(self):
        self.name = "public attribute"     #PUBLIC
        self._age = 21                     #PROTECTED
        self.__salary = 70000              #PRIVATE
    
    def show(self):
        print("inside the class")
        print("public:",self.name)
        print("protected:",self._age)
        print("private:",self.__salary)

obj = Demo()
obj.show()     #can't access age & salary
        
        
```
