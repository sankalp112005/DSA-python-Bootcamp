### CONSTRUCTOR
- constructor is a method that runs automatically when we call a class & this constructor function will target the object location.
- if we talk about function we can ask user using parameters,but in class we can't have parameter so we use constructor.
- to target object location we use 'self' keyword.

```python
class Factory:
    def __init__(self,material,zips,pocket):       #initialization function,we can ask user using ,
        self.material = material
        self.zips = zips
        self.pocket = pocket
    
    def show(self):
        print(f'your object details are {self.material},{self.zips},{self.pocket} ')

reebok = Factory("leather",20,40)              #reebok becomes instance(object) ,we called factory class inside reebok variable
campus = Factory("nylon",10,20)

campus.show()
print(reebok.pocket)
print(campus.material)
```

#### TYPES OF ATTRIBUTES
1. CLASS ATTRIBUTE - Normal variable cretaed inside the class is a class attribute.

2. INSTANCE ATTRIBUTE - attribute cretaed using the instance like self.name,self.age known as instace attribute. 

- 'self' capture our instace or object

```python
class Animal:
    name = 'lion'        #CLASS ATTRIBUTE
    year = 2002

    def __init__(self,age):
        self.age = age   #INSTANCE ATTRIBUTE   

obj = Animal(1)
print(obj.name)
print (obj.year)

obj2 = Animal(24)
print(obj2.age)

```

#### TYPES OF METHODS 
1. INSTANCE METHOD - an instance method works with instance(object) of the class.this method can access and modify instance attribute.

2. CLASS METHOD - this method works with the class itself it will not target the instance(object). we have to use @classmethod decorator for creating the class method and it takes 'cls' as thier first parameter.

3. STATIC METHOD - this method dosen't access class or instance directly it also uses a decorator @staticmethod it just acts like a regular function placed inside a class.

- 'cls' target the class location whereas 'self' target the object location.

```python
class Animal:
    name = "lion"       

    def __init__(self,age):
        self.age = age   
    
    def show(self):      #instance method
        print(f"how are you , your age is {age.self}")

    @classmethod         #decorator  ,,,CLASS METHOD
    def hello(cls):           #cls target the class location
        print("how are you brother")

    @staticmethod       #STATIC METHOD
    def static():
        print("how are u")

obj = Animal (12)
obj.hello()
obj.static()


```
