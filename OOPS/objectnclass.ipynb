# OBJECT ORIENTED PROGRAMMING SYSTEM (OOPS)
- it is based on the concept of "objects" which can contain data(attribute) & code(method).
- Class, Object, Encapsulation, Polymorphism, Inheritance etc are concept of oops.
- it makes code reusable,help in security etc.

## 1.CLASSES - 
- class is like a blueprint or template for creating objects.
- it is like blueprint of the house,it deifne what a house must have (rooms,kitchen)but dosen't build house. an object is the actual house built using that blueprint.
- created with basic keyword 'class' and a name infront of it.
- there are two types of things inside the class-
  1. Attributes - variable defined inside the class are attribute.
  2. Methods - functions defined inside the class are Methods.

- class is intialized first time when we run a program.
- for accessing attribute and methods we need to first access the class.


```python
class Factory:
    a=12          #a is attribute (vairable defined inside the class)

    def hello(self):  #hello is method (function defined inside the class)
        print ("how are you!")

    print("hello")   # it run when class is initialized

print(Factory().a)   #calling attribute (accessing attribute)

Factory().hello()    #call method
```

## 2.OBJECTS -
- real world entity created from class, that contains data(attribute) and behaviour(function).
- two components of class is 
 1. Attribute - variables that store information about object.
 2. Methods - functions that define the behavior of object.

- object has all the powers of class.

```python
class Factory:
    a=12          #a is attribute (vairable defined inside the class)

    def hello(self):  #hello is method (function defined inside the class)
        print ("how are you!")

    print("hello")

obj = Factory()    #we call class in 'obj' variable which converted to object.
print(obj.a)
obj.hello()

#we can create multiple object with single/same class
obj2 = Factory()   #it also become object
obj3 = Factory()
```
