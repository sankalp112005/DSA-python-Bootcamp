## INHERITANCE
- It allows a class(child class) to inherit properties and behaviour(attributes and methods) from another class(parent class).
- benefits of inheritance are-
1. Code reusability
2. Organized structure
3. Easy to maintain and extend 

- inehrtited class has all the powers of parent class, means all the methods and attribute can be accessed by the instance of child class as well

- let i created a parent class with constructor function inside it and then this class is inherited by another class then construtor function of parent class will work for child class as well.

```python
class Factorymumbai:                  #PARENT CLASS//SUPER CLASS
    a = "i am a attribute mentioned inside the Factory"
    def hello(self):
        print('hello i am a method mentioned inside the mumbai Factory')

class Factorypune(Factorymumbai):     #CHILD CLASS/ SUB CLASS 
    pass

obj = Factorymumbai()     #calling super class
obj2 = Factorypune()      #calling sub class
print(obj.a)
obj2.hello()              #obj2 is object of factorypune but it has all the powers of facotymumbai also.
```

- if you want to add new parameter in child class,then you have to create constuctor function for your child class.
- but parameter initialized in parent class will be initilized using 'super()' function.
- super() function will target parent class.

```python
#constructor function in inheritance
class Animal:
    def __init__(self,name):         #CONSTRUCTOR FUNCTION/INITIALIZED
        self.name =  name
    
    def show(self):
        print(f'hello your name is {self.name}')

class Human(Animal):                 #Human class inherit Animal class
    def __init__(self,name,age):      #we want age also in human case only
        super().__init__(name)     #it initalize the name from parent class
        #super() function target thr parent class 
        self.age = age             #age alg se initilaize krdo

        #if we are adding some extra functionality in derived class then, we need show function in that class also
    
    def show(self):
        print(f'your name is {self.name} your and age is {self.age}')


animal1 = Animal('lion')               #instance of main class
person1 = Human('sankalp',21)          #instance of child class

person1.show()
animal1.show()

```

#### TYPES OF INHERITANCE
1. SINGLE INHERITANCE - normal wla

2. MULTIPLE INHERITANCE -
 - there are 2 parent classes and only 1 child class 
 - that child class will inherit all the attributes and method of both parent class.

3. MULTILEVEL INHERITANCE - attribute and methods are passed on thorugh all the classes.

4. HIERARCHIAL INHERITANCE - it has 1 main class and 2 or more child class derived form main class.

```python
#MULTIPLE INHERITANCE
class Animal:
    name1 = 'lion'
class Human:
    name2 = 'harsh'
class Robots(Animal,Human):    #multiple inheritance
    name3 = "charlie11"

obj = Robots()
obj.name2
```

```python
class Animal:
    def __init__(self,name):
        pass
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age 
    def show(self):
        print(f'name {self.name} and age is {self.age}')
    #constructor function will we inherited of the first class that has been inherited.
    #this is known as MRO (Method Resolution Order) in python
class Robots(Human,Animal):    #here constructor function of human class will we inherited.(written first)
    name3 = "charlie11"

obj = Robots("SANKALP",21) 
obj.show()

```

###### MULTILEVEL INHERITANCE

```python
class Factory:                              #GRAND-PARENT CLASS
    def __init__(self,material,zips):       #material & zips will be initialize in this class
        self.material = material
        self.zips = zips

class BhopalFactory(Factory):               #PARENT CLASS
    def __init__(self,material,zips,color): #color is initialize in this class
        super().__init__(material,zips)
        self.color = color

class PuneFactory(BhopalFactory):                     #CHILD CLASS
    def __init__(self,material,zips,color,pockets):   #only pockets is intialized in this class
        super().__init__(material,zips,color)
        self.pockets = pockets

obj = PuneFactory("leather",2,'black',3)      #it takes 4 input 
obj2 = BhopalFactory("chelsi",4,"blue")       #it takes 3 argument
```
