```python
##BOTH MIX
def print_details(*args,**kwargs):
    for val in args:
        print(f"psoitional arguments {val}")

    for key,value in kwargs.items():
        print(f"KEYword arguments are {key} : {value}")

print_details(1,2,4,7,9,name = "sankalp",age = 21, college= "LNCT",city = "vidisha")

```

#### 1. TEMPERATURE CONVERSION

```python
def convert_temperature(temp,unit):
    ''' this fucntion converts temperature between celcius to faranheit'''
    if unit == 'C':
        return temp *9/5 + 32 ##Celcius to farenheit
    elif unit == 'F':
        return (temp-32)*5/9  ##Fahrehiet to clcius
    else:
        return None

print(convert_temperature(25,'C'))
print(convert_temperature(45,'F'))
    
```

##### PASSWORD STREGHT CHECKERR

```python
def strong_passsword(password):
    '''this funciton checks the streght of passowrd'''
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):   # start condition then iteration inn each charater
        return False
    if not any(char.islower() for char in password):       #char.islower() function
        return False
    if not any(char.isupper() for char in password):
        return False
    return True

#calling the function
print(strong_passsword("HELLO"))
print(strong_passsword("Hellosankalp123"))
```

#### 3.CALCULATE THE TOTAL COST OF ITEM IN SHOPPING CART'

```python
def calulate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost += item['price']* item['quantity']
    return total_cost

##Example cart data
cart = [
    {'name' : 'apple',"price": 0.5,'quantity':4},
    {'name':'banana', 'price': 0.3, 'quantity':6},
    {'name':'orange','price':0.7,'quantity': 3}
]

##CALIING THE FUCTION
total_cost = calulate_total_cost(cart)
print(total_cost)
```

#### 4.CHECK IF STRING IS PALINDROME OR NOT


```python
def is_palindrome(s):
    s=s.lower().replace(' ','')  #lower the string and remove the fre space
    return s==s[::-1]

print(is_palindrome('naman'))
print(is_palindrome('a man a plan a canal panama'))
```

#### 5.FACTORIAL OF NUMBER USING RECUSRION


```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(5))
```
